#### I.2 Biểu đồ thanh

```python
# Biểu đồ thanh từ một bảng tần số cho trước
import numpy as np
# nhãn của các thanh
TL=np.array(['R&B','Rock','Rap','Country','Classical','Latin']) 
# cao độ của các thanh (tần số của các biểu hiện)
SL=np.array([146.4, 102.6, 73.7,64.5, 14.8, 14.5])
# màu của các thanh, có thể 1 hoặc nhiều màu tùy ý
col=['blue', 'yellow']
# lệnh vẽ biểu đồ thanh
plt.bar(TL,SL,color = col)
plt.show()
```

```python
plt.bar # các tham số, chủ yếu dùng x, height, color
# x: nhãn của các thanh
# height: cao độ của các thanh (tần số của các biểu hiện)
# color: màu của các thanh
```

```python
?plt.bar 
# chi tiết hơn về các tham số
```

#### I.3 Biểu đồ tròn

```python
#b. Biểu đồ tròn
plt.pie(SL, labels=TL) # SL và TL được nhập số liệu ở bên trên
plt.show()
```

```python
plt.pie #các tham số, chủ yếu dùng x, labels, colors, cách dùng các tham số này
# giống trong plt.bar
```

### II. Dùng pandas

```python
Solieu['GioiTinh'].value_counts()
```

Để vẽ biểu đồ thanh cho biến giới tính, ta có thể dùng tần số tính được ở trên để vẽ trong matplotlib. Ngoài ra, pandas hỗ trợ vẽ với dữ liệu thô như sau.

```python
Solieu['GioiTinh'].value_counts().plot(kind='bar') # Dùng pandas
```
