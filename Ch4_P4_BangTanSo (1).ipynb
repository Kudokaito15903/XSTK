## 1. Tính số học sinh nữ, nam.

Muốn tính số học sinh nữ, nam ta sẽ lập bảng tần số cho biến Giới tính

```python
# import dữ liệu từ file csv: Dùng method pandas.read_csv()
import pandas as pd
sv=pd.read_csv('HocSinh.csv')
print(sv)
```

## 2. Tính số học sinh nam có mức độ yêu thích thể thao là 4.

Ta tìm kết quả bằng cách lập bảng tần số chéo của 2 biến là giới tính và thể thao.

```python
pd.crosstab(sv.GioiTinh,sv.TheThao)
```

Vậy có 9 học sinh nam có mức độ yêu thích thể thao là 4.

## 3. Tính tỷ lệ học sinh có mức độ yêu thích thể thao là 4.

Để tính tỷ lệ này ta lập tần suất cho biến thể thao

```python
sv.TheThao.value_counts(normalize = True)
# đặt normalize = True để tính tần suất
```

Vậy tỉ lệ cần tìm là 0.096842 = 9.6842%

#### 3.1 Tính tỷ lệ học sinh từ 9 tuổi trở lên

```python
tuoi9 = sv.Tuoi >= 9
tuoi9.value_counts(normalize=True)
```

## 4. Tính tỷ lệ học sinh có mức độ yêu thích thể thao là 4 trong nhóm học sinh nam, trong nhóm học sinh nữ.

Để tìm các tỷ lệ này ta tính tần suất của bảng tần số chéo của 2 biến giới tính và thể thao. Đây là bảng tần số chéo đó.

```python
pd.crosstab(sv.TheThao,sv.GioiTinh,margins=True)
# đặt margins=True để tính tổng cột và tổng hàng
```

Có 3 cách tính tần suất cho bảng này, được quy định bởi tham số normalize

 1. Tính tần suất đối với toàn bộ tập dữ liệu: normalize = 'all'
        
 2. Tính tần suất theo dòng: normalize = 'index'
    
 3. Tính tần suất theo cột: normalize = 'columns'

 Để tính tỷ lệ học sinh có mức độ yêu thích thể thao là 4 trong nhóm
học sinh nam, trong nhóm học sinh nữ ta tính tần suất theo cột, vì nam và nữ là 2 cột trong bảng tần số chéo trên.

```python
pd.crosstab(sv.TheThao,sv.GioiTinh,normalize='columns')
```

Tỷ lệ học sinh có mức độ yêu thích thể thao là 4 trong nhóm
học sinh nam, trong nhóm học sinh nữ lần lượt là 0.039648 và 0.149194

#### 4.1 Tính tỷ lệ học sinh từ 9 tuổi trở lên trong nhóm nam

```python
pd.crosstab(tuoi9,sv.GioiTinh,normalize='columns') # tuoi9 đã tạo ra trong mục 3.1
```

## 5. Tính tỉ lệ những học sinh là nam và có mức độ yêu thích thể thao bằng 4

```python
pd.crosstab(sv.TheThao,sv.GioiTinh,normalize='all')
```

Vậy tỉ lệ những học sinh là nam và có mức độ yêu thích thể thao bằng 4 là 0.018947

1. Tiến hành phân tổ cột dữ liệu về tuổi thành các tổ với các điểm chia là 20,30,40,50,60,70,80
và tính tỉ lệ những người được điều tra có độ tuổi không vượt quá 50.

```python
sl=pd.read_csv('SoLieu.csv')
```

Các bước Phân tổ dữ liệu:
    
1. Tìm min, max

2. Xác định các điểm chia (Chú ý: điểm chia nhỏ nhất <= min <= max <= điểm chia lớn nhât)

3. Dùng lệnh pd.cut để phân tổ dữ liệu

```python
min(sl.Tuoi)
```

```python
max(sl.Tuoi)
```

```python
?pd.cut
```

```python
ptTuoi=pd.cut(sl.Tuoi,bins=[20, 30, 40, 50, 60, 70, 80],
              include_lowest=True)
# bins: các điểm chia
# include_lowest = True: làm cận dưới của tổ đầu tiên giảm đi
# một ít (để chứa cả giá trị 20)
```

```python
# tính tỉ lệ những người được điều tra có độ tuổi không vượt quá 50.
ptTuoi.value_counts(sort = False, normalize = True).cumsum()
# sort = False để xếp các tổ theo chiều tăng của dữ liệu,
# ngược lại theo chiều giảm của tần số (tần suất)
# cumsum để tính tần suất tích lũy
```

Vậy tỉ lệ này là 73%
